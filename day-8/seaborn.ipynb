```python

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

data = {
    'Name': ['Alice', 'Bob', 'Charlie', 'David', 'Eve'],
    'Age': [25, 30, 50, 40, 45],
}

df = pd.DataFrame(data)

sns.lineplot(x = df.index, y= 'Age',data=df)

plt.title('Age of Individuals')
plt.xlabel('Index')
plt.ylabel('Age')
plt.show()

```
