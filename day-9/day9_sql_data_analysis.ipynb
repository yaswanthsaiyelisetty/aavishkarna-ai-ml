# Day 9: SQL + Data Analysis using Python, Pandas, and SQLite

This notebook demonstrates beginner-level data analysis by loading a dataset, inserting it into a local SQLite database, and running basic SQL queries using Pandas.

## 1. Import Libraries
We need `pandas` for data manipulation and `sqlite3` to interact with our local database.

````python
import pandas as pd
import sqlite3
````

## 2. Load Dataset
Read the CSV file into a pandas DataFrame.

````python
df = pd.read_csv("products.csv")
````

## 3. Display Basic Information
Check the first 5 rows, the dataset shape (number of rows and columns), and the column names.

````python
print("Shape of dataset:", df.shape)
print("\nColumn names:", df.columns.tolist())

print("\nFirst 5 rows:")
df.head()
````

## 4. Create SQLite Database and Table
We establish a connection to a database file named `products.db` (it will be created if it doesn't exist). Then, we store our pandas DataFrame into a SQL table named `products`.

````python
conn = sqlite3.connect("products.db")
df.to_sql("products", conn, if_exists="replace", index=False)
print("Data successfully stored in 'products.db' database under the 'products' table.")
````

## 5. Execute SQL Queries

### a) View first 5 records
This query retrieves all columns for the first 5 records from the `products` table.
```sql
SELECT * FROM products LIMIT 5;
```

````python
query_a = "SELECT * FROM products LIMIT 5;"
pd.read_sql(query_a, conn)
````

### b) Select specific columns
This query selects only the `Name`, `Category`, and `Price` columns (showing just the first 5 results here for brevity).
```sql
SELECT Name, Category, Price
FROM products;
```

````python
query_b = """
SELECT Name, Category, Price
FROM products;
"""
pd.read_sql(query_b, conn).head()
````

### c) Use WHERE clause
This query filters the dataset to only include products with a price strictly greater than 200.
```sql
SELECT Name, Category, Price
FROM products
WHERE Price > 200;
```

````python
query_c = """
SELECT Name, Category, Price
FROM products
WHERE Price > 200;
"""
pd.read_sql(query_c, conn).head()
````

### d) Use ORDER BY
This query retrieves the names, categories, and prices of products, sorting them by price in descending order (highest price first).
```sql
SELECT Name, Category, Price
FROM products
ORDER BY Price DESC;
```

````python
query_d = """
SELECT Name, Category, Price
FROM products
ORDER BY Price DESC;
"""
pd.read_sql(query_d, conn).head()
````

### e) Use GROUP BY with COUNT
This query groups the products by category and counts how many products belong to each category.
```sql
SELECT Category,
       COUNT(*) AS Product_Count
FROM products
GROUP BY Category;
```

````python
query_e = """
SELECT Category,
       COUNT(*) AS Product_Count
FROM products
GROUP BY Category;
"""
pd.read_sql(query_e, conn).head()
````

### f) Use GROUP BY with AVG
This query calculates the average price per category and sorts the results so the category with the highest average price appears first.
```sql
SELECT Category,
       AVG(Price) AS Average_Price
FROM products
GROUP BY Category
ORDER BY Average_Price DESC;
```

````python
query_f = """
SELECT Category,
       AVG(Price) AS Average_Price
FROM products
GROUP BY Category
ORDER BY Average_Price DESC;
"""
pd.read_sql(query_f, conn).head()
````

## 6. Close the Database Connection
It is a best practice to close the connection to the database once you are done.

````python
conn.close()
print("Database connection closed.")
````
