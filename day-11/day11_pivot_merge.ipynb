# Day 11: Pivot and Merge Activity

In this notebook, we perform basic data reshaping and combining using Pandas `pivot_table` and `merge` functions.

## 1. Import Pandas and Load Data

```python
import pandas as pd
import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv('products.csv')

# Fix the 'Price' column which might contain strings or commas from the raw data
df['Price'] = pd.to_numeric(df['Price'], errors='coerce')
```

## 2. Basic Dataset Information

```python
print("Shape:", df.shape)
print("\nColumns:", df.columns.tolist())
print("\nFirst 5 rows:")
df.head()
```

```python
df.info()
```

---
## Part 1: Pivot Activity

**Explanation**: The pivot table below summarizes the average (mean) product price for each `Category`, broken down by product `Availability`.

```python
pivot_table = pd.pivot_table(
    df,
    index="Category",
    columns="Availability",
    values="Price",
    aggfunc="mean"
)

pivot_table.head()
```

---
## Part 2: Merge Activity

Since we only have one dataset, we will create a manual `discount_df` to simulate joining two data sources.

```python
discount_df = pd.DataFrame({
    "Category": ["Books & Stationery", "Kitchen Appliances", "Shoes & Footwear", "Kids' Clothing"],
    "Discount_Percentage": [10, 15, 20, 5]
})

merged_df = pd.merge(df, discount_df, on="Category", how="left")
merged_df.head(10)
```

---
## Part 3: Final Output

We can display the final combined dataset and save it to a new CSV file.

```python
merged_df.to_csv("day11_merged_products.csv", index=False)
print("Final merged DataFrame saved as 'day11_merged_products.csv'!")

merged_df.head()
```

---
## Conclusion
- **Pivot Table**: Used to summarize product prices category-wise. It is great for restructuring data to see multi-dimensional aggregations (like category vs availability).
- **Merge**: Used to combine product data with discount data based on the shared `Category` column. This is essential for joining related datasets from different sources.
- Both operations are foundational tools in data analysis.
