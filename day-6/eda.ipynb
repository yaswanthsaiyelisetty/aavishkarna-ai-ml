# Exploratory Data Analysis (EDA) on Products Dataset

This notebook contains a complete step-by-step Exploratory Data Analysis (EDA) of the `products.csv` dataset, implementing steps 1 to 8 of the EDA lifecycle.

## Step 1 & 2: Load the Dataset and Set up the Environment

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Set plotting style
sns.set_theme(style="whitegrid")
plt.rcParams["figure.figsize"] = (10, 6)

# Load products.csv dataset
df = pd.read_csv('products.csv')

# Output basic information
print(f"Dataset Shape: {df.shape[0]} rows, {df.shape[1]} columns")
df.head()
```

## Step 3: Data Inspection & Summary Statistics

```python
# Inspect column data types and non-null values
print("=== Data Schema & Summary ===")
df.info()

print("\n=== Descriptive Statistics for Numerical Features ===")
display(df.describe())

print("\n=== Descriptive Statistics for Categorical Features ===")
display(df.describe(include=['object']))
```

## Step 4 & 5: Data Cleaning & Handling Duplicates / Anomalies

```python
# 1. Check for duplicate rows
duplicate_count = df.duplicated().sum()
print(f"Number of duplicate rows: {duplicate_count}")

# 2. Check for missing values
missing_values = df.isnull().sum()
print("\n=== Missing Values Count ===")
print(missing_values[missing_values > 0] if missing_values.sum() > 0 else "No missing values found!")

# 3. Value-range checks for Price and Stock to ensure no anomalous negative numbers
negative_prices = (df['Price'] < 0).sum()
negative_stock = (df['Stock'] < 0).sum()
print(f"\nNegative Prices found: {negative_prices}")
print(f"Negative Stock count: {negative_stock}")

# 4. Strip any potential leading/trailing whitespaces in string columns
string_cols = df.select_dtypes(include=['object']).columns
for col in string_cols:
    df[col] = df[col].astype(str).str.strip()

print("\nData cleaning steps completed successfully.")
```

## Step 6: Univariate Analysis

```python
# 1. Distribution of Product Prices
plt.figure(figsize=(10, 4))
sns.histplot(df['Price'], kde=True, color='teal', bins=30)
plt.title('Distribution of Product Prices', fontsize=14)
plt.xlabel('Price ($)')
plt.ylabel('Count')
plt.show()

# 2. Distribution of Product Stock Quantity
plt.figure(figsize=(10, 4))
sns.histplot(df['Stock'], kde=True, color='darkorange', bins=30)
plt.title('Distribution of Product Stock Quantities', fontsize=14)
plt.xlabel('Stock Quantity')
plt.ylabel('Count')
plt.show()

# 3. Top 10 Product Categories
plt.figure(figsize=(10, 5))
top_categories = df['Category'].value_counts().head(10)
sns.barplot(x=top_categories.values, y=top_categories.index, palette='viridis')
plt.title('Top 10 Product Categories by Count', fontsize=14)
plt.xlabel('Product Count')
plt.ylabel('Category')
plt.show()

# 4. Distribution of Product Availability Status
plt.figure(figsize=(8, 4))
sns.countplot(data=df, x='Availability', palette='pastel', order=df['Availability'].value_counts().index)
plt.title('Product Count by Availability Status', fontsize=14)
plt.xlabel('Availability')
plt.ylabel('Count')
plt.show()
```

## Step 7 & 8: Bivariate / Advanced Analysis

```python
# 1. Correlation Matrix of Numerical Features
plt.figure(figsize=(6, 4))
correlation = df[['Price', 'Stock', 'Internal ID']].corr()
sns.heatmap(correlation, annot=True, cmap='coolwarm', vmin=-1, vmax=1, fmt=".2f")
plt.title('Correlation Matrix of Numerical Features', fontsize=13)
plt.show()

# 2. Relationship between Price and Stock Quantity
plt.figure(figsize=(10, 5))
sns.scatterplot(data=df, x='Price', y='Stock', alpha=0.6, color='purple')
plt.title('Product Price vs. Stock Quantity', fontsize=14)
plt.xlabel('Price ($)')
plt.ylabel('Stock')
plt.show()

# 3. Average Price by Availability Status
plt.figure(figsize=(8, 4))
sns.barplot(data=df, x='Availability', y='Price', palette='muted', errorbar=None)
plt.title('Average Product Price by Availability Status', fontsize=14)
plt.xlabel('Availability')
plt.ylabel('Average Price ($)')
plt.show()

# 4. Boxplot of Prices for the Top 5 Categories
plt.figure(figsize=(10, 5))
top_5_cats = df['Category'].value_counts().head(5).index
df_top_5 = df[df['Category'].isin(top_5_cats)]
sns.boxplot(data=df_top_5, x='Price', y='Category', palette='Set3')
plt.title('Price Distribution across Top 5 Product Categories', fontsize=14)
plt.xlabel('Price ($)')
plt.ylabel('Category')
plt.show()
```
