### 01. Basic Line Chart

```python
import matplotlib.pyplot as plt
import numpy as np

# Data preparation
x = np.array([1, 2, 3, 4, 5])
y = np.array([2, 4, 6, 8, 10])

# Plotting
plt.plot(x, y)

# Customizations
plt.title("Basic Line Chart")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")

# Show plot
plt.show()
```

### 02. Basic Bar Chart

```python
import matplotlib.pyplot as plt

# Data preparation
categories = ['A', 'B', 'C', 'D', 'E']
values = [5, 7, 3, 8, 6]

# Plotting
plt.bar(categories, values)

# Customizations
plt.title("Basic Bar Chart")
plt.xlabel("Categories")
plt.ylabel("Values")

# Show plot
plt.show()
```

### 03. Basic Histogram

```python
import matplotlib.pyplot as plt
import numpy as np

# Generate random data
np.random.seed(42)
data = np.random.randn(1000)

# Plotting
plt.hist(data, bins=30)

# Customizations
plt.title("Basic Histogram")
plt.xlabel("Value Intervals")
plt.ylabel("Frequency")

# Show plot
plt.show()
```

### 04. Basic Scatter Plot

```python
import matplotlib.pyplot as plt
import numpy as np

# Data preparation
np.random.seed(42)
x = np.random.rand(50) * 10
y = np.random.rand(50) * 10

# Plotting
plt.scatter(x, y)

# Customizations
plt.title("Basic Scatter Plot")
plt.xlabel("X-axis Values")
plt.ylabel("Y-axis Values")

# Show plot
plt.show()
```

### 05. Basic Pie Chart

```python
import matplotlib.pyplot as plt

# Data preparation
labels = ['Apples', 'Bananas', 'Cherries', 'Dates']
sizes = [35, 25, 25, 15]

# Plotting
plt.pie(sizes, labels=labels)

# Customizations
plt.title("Basic Pie Chart")

# Show plot
plt.show()
```

### 06. Basic Box Plot

```python
import matplotlib.pyplot as plt
import numpy as np

# Generate random data representing multiple groups
np.random.seed(42)
data_group1 = np.random.normal(100, 10, 200)
data_group2 = np.random.normal(90, 20, 200)
data_group3 = np.random.normal(110, 15, 200)

data_to_plot = [data_group1, data_group2, data_group3]

# Plotting
plt.boxplot(data_to_plot, labels=['Group 1', 'Group 2', 'Group 3'])

# Customizations
plt.title("Basic Box Plot")
plt.ylabel("Value Distribution")

# Show plot
plt.show()
```

### 07. Basic Heatmap

```python
import matplotlib.pyplot as plt
import numpy as np

# Generate a 2D random grid
np.random.seed(42)
data_matrix = np.random.rand(10, 10)

# Plotting heatmap
plt.imshow(data_matrix, cmap='viridis', interpolation='nearest')

# Add colorbar indicator
plt.colorbar(label="Density Index")

# Customizations
plt.title("Basic Heatmap")
plt.xlabel("X Coordinate")
plt.ylabel("Y Coordinate")

# Show plot
plt.show()
```

### 08. Custom Line Chart (with custom styles, markers, and gridlines)

```python
import matplotlib.pyplot as plt
import numpy as np

# Data preparation
x = np.arange(0, 10, 1)
y = x ** 1.5

# Plotting customized line
plt.plot(x, y, color='red', linestyle='--', marker='o', linewidth=2.5, markersize=8, label="Trend Line")

# Customizations
plt.title("Line Chart with custom styles, markers, and gridlines")
plt.xlabel("Time Duration")
plt.ylabel("Growth Scale")
plt.grid(True, which='both', linestyle=':', color='gray', alpha=0.7)
plt.legend(loc="upper left")

# Show plot
plt.show()
```

### 09. Custom Bar Chart (with customized colors and borders)

```python
import matplotlib.pyplot as plt

# Data preparation
labels = ['Product A', 'Product B', 'Product C', 'Product D']
sales = [120, 180, 90, 210]
custom_colors = ['#1f77b4', '#ff7f0e', '#2ca02c', '#d62728']

# Plotting customized bar
plt.bar(labels, sales, color=custom_colors, edgecolor='darkblue', linewidth=2.0)

# Customizations
plt.title("Bar Chart with customized colors and borders")
plt.xlabel("Product Types")
plt.ylabel("Sales Volume")

# Show plot
plt.show()
```

### 10. Custom Histogram (with custom bin spacing and transparency)

```python
import matplotlib.pyplot as plt
import numpy as np

# Generate data
np.random.seed(42)
scores = np.random.normal(75, 10, 500)

# Define customized bin spacing
custom_bins = [40, 50, 60, 70, 80, 90, 100]

# Plotting customized histogram
plt.hist(scores, bins=custom_bins, color='mediumseagreen', alpha=0.6, edgecolor='black', facecolor='mediumseagreen')

# Customizations
plt.title("Histogram with custom bin spacing and transparency")
plt.xlabel("Score Intervals")
plt.ylabel("Student Count")

# Show plot
plt.show()
```

### 11. Custom Scatter Plot (with custom square markers and colors)

```python
import matplotlib.pyplot as plt
import numpy as np

# Generate coordinates
np.random.seed(42)
x = np.random.rand(40) * 100
y = np.random.rand(40) * 100

# Plotting customized scatter
plt.scatter(x, y, color='purple', marker='s', s=120, alpha=0.75, edgecolor='black', linewidths=1.2)

# Customizations
plt.title("Scatter Plot with custom square markers and colors")
plt.xlabel("Dimension X")
plt.ylabel("Dimension Y")

# Show plot
plt.show()
```

### 12. Custom Pie Chart (with shadow effects and exploded slices)

```python
import matplotlib.pyplot as plt

# Data preparation
labels = ['Marketing', 'Research & Dev', 'Operations', 'Finance']
budget_shares = [30, 40, 20, 10]

# Explosion array to separate the 'R&D' slice
explode_shares = [0, 0.1, 0, 0]
custom_colors = ['#ff9999', '#66b3ff', '#99ff99', '#ffcc99']

# Plotting customized pie
plt.pie(budget_shares, explode=explode_shares, labels=labels, autopct='%1.1f%%', shadow=True, startangle=140, colors=custom_colors)

# Customizations
plt.title("Pie Chart with shadow effects and exploded slices")

# Show plot
plt.show()
```
