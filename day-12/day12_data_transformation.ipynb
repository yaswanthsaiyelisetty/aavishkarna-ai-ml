## Program 1: Handling Missing Values

- **Missing values**: Occur when no data value is stored for an observation. They can cause errors in machine learning models.
- **Mean imputation**: Fills missing values with the average of the numerical column.
- **Median imputation**: Fills missing values with the middle value (less sensitive to outliers).
- **Mode imputation**: Fills missing values with the most frequent value (used for categorical data or discrete numbers).

```python
import pandas as pd
import numpy as np
import warnings
warnings.filterwarnings('ignore')

# Create sample DataFrame with missing values
data = {'Age': [25, np.nan, 22, 29, 29],
        'Salary': [50000, 54000, np.nan, 58000, 58000]}
df_missing = pd.DataFrame(data)

print("Original Data with missing values:")
print(df_missing)

# Fill using mean
df_mean = df_missing.copy()
df_mean['Age'] = df_mean['Age'].fillna(df_mean['Age'].mean())

# Fill using median
df_median = df_missing.copy()
df_median['Salary'] = df_median['Salary'].fillna(df_median['Salary'].median())

# Fill using mode
df_mode = df_missing.copy()
df_mode['Age'] = df_mode['Age'].fillna(df_mode['Age'].mode()[0])

print("\nAfter Mean Imputation (Age):")
print(df_mean)
print("\nAfter Median Imputation (Salary):")
print(df_median)
print("\nAfter Mode Imputation (Age):")
print(df_mode)
```

## Program 2: Handling Outliers using IQR

- **Outliers**: Data points that differ significantly from other observations. They can skew our analysis.
- **IQR method**: Interquartile Range (IQR = Q3 - Q1) is used to find the spread of the middle 50% of data. Values outside (Q1 - 1.5 * IQR) and (Q3 + 1.5 * IQR) are considered outliers.

```python
# Create sample dataset with outliers
data = {'Scores': [55, 60, 65, 70, 75, 80, 250, 300]} # 250 and 300 are outliers
df_outliers = pd.DataFrame(data)

print("Original Data:")
print(df_outliers)

# Detect outliers
Q1 = df_outliers['Scores'].quantile(0.25)
Q3 = df_outliers['Scores'].quantile(0.75)
IQR = Q3 - Q1
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Remove outliers
df_cleaned = df_outliers[(df_outliers['Scores'] >= lower_bound) & (df_outliers['Scores'] <= upper_bound)]

print(f"\nBounds: Lower={lower_bound}, Upper={upper_bound}")
print("\nCleaned Data (Without Outliers):")
print(df_cleaned)
```

## Program 3: Normalization using MinMaxScaler

- **Normalization**: Scales the values of a column to a specific range, usually between 0 and 1, without changing the shape of the original distribution.

```python
from sklearn.preprocessing import MinMaxScaler

# Create sample dataset
data = {'Price': [100, 200, 300, 400, 500]}
df_norm = pd.DataFrame(data)

print("Original Data:")
print(df_norm)

# Apply MinMaxScaler
scaler = MinMaxScaler()
df_norm['Price_Scaled'] = scaler.fit_transform(df_norm[['Price']])

print("\nData after MinMaxScaler:")
print(df_norm)
```

---
## Program 4: Standardization using StandardScaler

- **Standardization**: Centers the data around a mean of 0 and a standard deviation of 1. It is very useful for models that assume normally distributed data (like SVM, Logistic Regression).

```python
from sklearn.preprocessing import StandardScaler

# Create sample dataset
data = {'Weight': [50, 60, 70, 80, 90]}
df_std = pd.DataFrame(data)

print("Original Data:")
print(df_std)

# Apply StandardScaler
std_scaler = StandardScaler()
df_std['Weight_Scaled'] = std_scaler.fit_transform(df_std[['Weight']])

print("\nData after StandardScaler:")
print(df_std)
```

---
## Program 5: Label Encoding

- **Label Encoding**: Converts categorical text data into numerical format, assigning a unique integer to each distinct category.

```python
from sklearn.preprocessing import LabelEncoder

# Create categorical dataset
data = {'Size': ['Small', 'Medium', 'Large', 'Medium', 'Small']}
df_label = pd.DataFrame(data)

print("Original Data:")
print(df_label)

# Apply LabelEncoder
label_enc = LabelEncoder()
df_label['Size_Encoded'] = label_enc.fit_transform(df_label['Size'])

print("\nData after Label Encoding:")
print(df_label)
```

---
## Program 6: One Hot Encoding

- **One-hot encoding**: Converts categorical variables into a set of binary columns (0s and 1s), preventing the model from assuming a numerical hierarchy between categories.

```python
# Create categorical dataset
data = {'Color': ['Red', 'Blue', 'Green', 'Red']}
df_ohe = pd.DataFrame(data)

print("Original Data:")
print(df_ohe)

# Apply get_dummies()
df_ohe_encoded = pd.get_dummies(df_ohe, columns=['Color'])

print("\nData after One-Hot Encoding:")
print(df_ohe_encoded)
```

## Program 7: Feature Engineering

- **Feature Engineering**: Creating new meaningful features from existing data to help machine learning models learn better patterns.

```python
# Create Length and Width columns
data = {'Length': [10, 15, 20],
        'Width': [5, 10, 15]}
df_feat = pd.DataFrame(data)

print("Original Data:")
print(df_feat)

# Create Area = Length x Width
df_feat['Area'] = df_feat['Length'] * df_feat['Width']

print("\nUpdated Dataset with new feature (Area):")
print(df_feat)
```

## Program 8: Log Transformation

- **Skewed Data**: When data is concentrated on one side and has a long tail on the other side.
- **Log Transformation**: Applies a logarithm to the data to reduce the impact of extreme values (outliers) and make highly skewed distributions more normally distributed.

```python
# Create skewed numerical data
data = {'Income': [3000, 4000, 5000, 100000, 500000]} # High skew due to large incomes
df_log = pd.DataFrame(data)

print("Original Skewed Data:")
print(df_log)

# Apply np.log1p() which calculates log(1+x) safely
df_log['Income_Log'] = np.log1p(df_log['Income'])

print("\nData after Log Transformation:")
print(df_log)
```

---
## Program 9: GroupBy Aggregation

- **groupby()**: Splits the data into groups based on some criteria, applies a function (like sum or mean) to each group, and combines the results.

```python
# Create sales dataset
data = {'Category': ['Electronics', 'Clothing', 'Electronics', 'Clothing', 'Electronics'],
        'Sales': [200, 50, 300, 80, 150]}
df_sales = pd.DataFrame(data)

print("Original Data:")
print(df_sales)

# Calculate category-wise total sales
total_sales = df_sales.groupby('Category')['Sales'].sum().reset_index(name='Total_Sales')

# Calculate average sales
avg_sales = df_sales.groupby('Category')['Sales'].mean().reset_index(name='Average_Sales')

print("\nTotal Sales by Category:")
print(total_sales)

print("\nAverage Sales by Category:")
print(avg_sales)
```

## Program 10: Complete Data Transformation Pipeline

- **Complete Transformation Workflow**: Real-world data is messy and usually requires a combination of operations: handling missing values, encoding text, and scaling numbers.

```python
# Create dataset containing missing values, categorical, and numerical values
data = {
    'City': ['New York', 'Paris', 'London', 'Paris', 'New York'],
    'Age': [25, np.nan, 30, 22, 28],
    'Salary': [60000, 50000, 80000, 45000, np.nan]
}
df_pipeline = pd.DataFrame(data)

print("Original Dirty Dataset:")
print(df_pipeline)

# 1. Missing value handling (Mean imputation for Age and Salary)
df_pipeline['Age'] = df_pipeline['Age'].fillna(df_pipeline['Age'].mean())
df_pipeline['Salary'] = df_pipeline['Salary'].fillna(df_pipeline['Salary'].mean())

# 2. Label Encoding (for City)
le = LabelEncoder()
df_pipeline['City_Encoded'] = le.fit_transform(df_pipeline['City'])

# 3. Normalization (MinMaxScaler for Salary)
scaler_pipe = MinMaxScaler()
df_pipeline['Salary_Scaled'] = scaler_pipe.fit_transform(df_pipeline[['Salary']])

print("\nFinal Transformed Dataset:")
print(df_pipeline)
```

## Final Conclusion

- **Importance of data transformation**: Raw data is rarely ready for machine learning. Transformation ensures the data is in the right format, range, and quality.
- **Importance of handling missing values**: ML algorithms usually crash or learn poorly if missing data (NaN) is left untreated.
- **Importance of scaling data**: Scaling (Normalization/Standardization) prevents features with larger numbers (like Salary=60000) from dominating features with smaller numbers (like Age=25).
- **Importance of encoding categorical variables**: ML models only understand numbers, so text like 'Red' or 'Blue' must be converted into numerical formats.
- **How these transformations help machine learning models perform better**: Properly transformed data leads to faster training, better accuracy, and a more robust predictive model.
